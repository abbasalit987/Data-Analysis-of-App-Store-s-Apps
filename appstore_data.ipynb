# Data Visualisation

```python
grp = appstore_data.groupby('prime_genre')
x = grp['user_rating'].agg(np.mean)
y = grp['price'].agg(np.sum)
print(x)
print(y)
```

```python
plt.plot(x,'ro')
plt.xticks(rotation =90)
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(x, 'ro', color = 'r')
plt.xticks(rotation =90)
plt.title('Prime Genre wise Rating ')
plt.xlabel('Prime Genre -->')
plt.ylabel('Rating -->')

plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(y, 'r--', color = 'b')
plt.xticks(rotation =90)
plt.title('Prime Genre wise Pricing ')
plt.xlabel('Prime Genre -->')
plt.ylabel('Price -->')
plt.show()
```
